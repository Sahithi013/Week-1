```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("mostafaabla/garbage-classification")

print("Path to dataset files:", path)
```

```python
import os
import matplotlib.pyplot as plt


# 1. Set dataset path

dataset_path = "dataset/garbage_classification"  # folder containing class subfolders
print("Dataset path:", dataset_path)
```

```python
# 2. Explore dataset

classes = os.listdir(dataset_path)
print("Classes found:", classes)

# Count images per class
for cls in classes:
    class_path = os.path.join(dataset_path, cls)
    print(f"{cls}: {len(os.listdir(class_path))} images")

# Display sample images
plt.figure(figsize=(12,6))
for i, cls in enumerate(classes[:6]):  # show first 6 classes
    img_path = os.path.join(dataset_path, cls, os.listdir(os.path.join(dataset_path, cls))[0])
    img = plt.imread(img_path)
    plt.subplot(2, 3, i+1)
    plt.imshow(img)
    plt.title(cls)
    plt.axis('off')
plt.show()
```
